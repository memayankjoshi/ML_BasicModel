## Linear Regression

```python
#Housing Price dataset
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset.head()
```

```python
dataset["Price"]=df.target
dataset.head()
```

```python
# Dividing the dataset into independent and dependent feature
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

# Linear Regression
`lin_reg.predict(input)` is used for predictions

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X_train,y_train,scoring="neg_mean_squared_error",cv=10)
mse_mean=np.mean(mse)
print(mse_mean)
```

# Ridge Regression

```python
## Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,40,100]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
print(ridge_regressor.best_params_) # best parameter
```

```python
print(ridge_regressor.best_score_) # best score
```

```python
y_pred=ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## Lasso Regression


```python
## Laso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso(max_iter=5000)
params={'alpha':[1e-3,1e-2,1,5,10,20,30,40,50,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_) # best parameter
print(lasso_regressor.best_score_) # best score
```

```python
y_pred=lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## Logistic Regression


```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
X=pd.DataFrame(df["data"],columns=df["feature_names"])
y=pd.DataFrame(df["target"],columns=["Target"])
```

```python
X.head()
```

```python
# Checking balanced dataset or not
y["Target"].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
X_train.head()
```

```python
params = {
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [500, 1000, 2000]
}
model1=LogisticRegression()
grid_search=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
grid_search.fit(X_train, y_train.ravel())
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
y_pred=grid_search.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
